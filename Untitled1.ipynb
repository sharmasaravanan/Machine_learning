```python
count = 0
opt='y'
while opt=='y':
    num=int(input("Enter the value for num :"))
    if num==15:
        print("goodbye...")
        print(count)
        break
    elif num==25:
        print("Skipping...")
        continue
    elif num==10:
        count+=1
        pass
    if num%4==0:
        print("Divisible by 4 and also even")
    elif num%2==0:
        print("even")
    else:
        print("odd")
    opt = input("Do you want to continue? y/n :")
```
