```python
def gettingInput():
    a = int(input("a :"))
    b = int(input("b :"))
    return a,b

def prime(i):+
    flag = True
    for j in range(2,int(i/2)):
        if i%j == 0:
            flag = False
            break
    return flag
```

```python
try:
    a,b = gettingInput()
    if (b-a) <10 or len(str(a))!=2 or len(str(b))!=2 :
        raise ValueError("Condition not satisfied")
except ValueError:
    a,b = gettingInput()
    
ans = []
for i in range(a,b):
    flag = prime(i)
    if flag:
        if int(str(i)[0])+int(str(i)[1]) <=7:
            ans.append(i)
print(ans)
```

```python
class parent:       
    def show(self,):
        print(self.name)
        print(self.number)
    
    def fruit(self,):
        print(self.fruitName)
```

```python
class child(parent):
    def __init__(self,name1,number1,address1,fruitName):
        self.name = name1
        self.number = number1
        self.address = address1
        self.fruitName = fruitName
    
    def display(self,):
        parent.show(self,)
        print(self.address)
        parent.fruit(self,)
        
```

```python
child = child("sharma saravanan",7418490015,"chennai","watermelon")
```

```python
child.display()
```

```python
class A:
    def __init__(self, name):
        self.name = name

    def show(self):
        print('name: %s' % (self.name,))


class B(A):
    def __init__(self, name, desc):
        A.__init__(self, name)
        self.desc = desc

    def show(self):
        A.show(self)
        print('desc: %s' % (self.desc,))


f = B("apple", "mango")
f.show()
```
