```python
i = 1
while i <= 3:
    num = int(input("Enter the numerical value : "))
    if num % 4 == 0:
        print("Divisible by 4 and even")
    elif num % 2 == 0:
        print('even')
    else:
        print('odd')
    i += 1
else:
    print("3 iterations are over")
```
