```python
#inheritance
class parent:
    cost = 115
    def show(self,):
        print("You bought %s fruit"%self.fruit)
    def calculate(self,):
        self.total = self.quan * parent.cost
        print("you have to pay %0.2f"%self.total)
```

```python
class child(parent):
    def __init__(self,fruit,quan):
        self.fruit = fruit
        self.quan = quan
    def display(self):
        parent.show(self)
        parent.calculate(self)
```
